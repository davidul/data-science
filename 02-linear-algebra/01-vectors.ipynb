Broadcasting is a powerful feature in NumPy that allows arrays of different shapes to be used together in element-wise operations. It eliminates the need for explicit loops and greatly simplifies the code.

In NumPy, broadcasting follows a set of rules to determine how arrays with different shapes can be combined. The smaller array is "broadcasted" to match the shape of the larger array, so that they can be operated on element-wise.

The broadcasting rules are as follows:
1. If the arrays have different numbers of dimensions, the smaller array is padded with ones on its left side until the number of dimensions match.
2. If the shape of the arrays does not match in any dimension, the array with shape equal to 1 in that dimension is stretched to match the shape of the other array.
3. If the shape of the arrays does not match in any dimension and neither is equal to 1, an error is raised.

Broadcasting allows for efficient and concise code when working with arrays of different shapes. It enables operations like adding a scalar to a matrix, multiplying a matrix by a vector, or performing element-wise operations on arrays of different sizes.

Note that broadcasting does not create new copies of the arrays, but rather uses a "virtual" array that represents the result of the operation. This makes broadcasting memory-efficient.

Overall, broadcasting is a powerful tool in NumPy that simplifies array operations and improves code readability.

````python
v4 = np.array([1,2]) 
v5 = np.array([1])
v6 = v4 + v5 # v5 is broadcasted to [1,1]
````

````python
v6
````

In geometric terms, vector addition can be visualized as placing the tail of the second vector at the head of the first vector, then drawing a new vector from the tail of the first vector to the head of the second vector. This new vector represents the sum of the two vectors. This is often referred to as the "tip-to-tail" method of vector addition.


The vector `v4` can be represented as a point in 2D space at coordinates (1,2). The vector `v5` is a 1D vector, but due to broadcasting, it becomes a 2D vector with coordinates (1,1). When you add `v4` and `v5`, you're effectively moving the point represented by `v4` one unit along both the x and y axes. The resulting vector `v6` represents a point at coordinates (2,3).

````python
import numpy as np
import matplotlib.pyplot as plt

# redefine v1 and v2
v1 = np.array([3,-1])
v2 = np.array([2,4])
v3 = v1 + v2 # add v1 and v2 together element-wise
# v3 -> [3+2, -1+4] -> [5, 3]

plt.arrow(0,0,v1[0],v1[1],head_width=0.2,head_length=0.2,color='b') # plot v1 as a blue arrow
plt.arrow(0,0,v2[0],v2[1],head_width=0.2,head_length=0.2,color='r') # plot v2 as a red arrow
plt.arrow(0,0,v3[0],v3[1],head_width=0.2,head_length=0.2,color='k') # plot v3 as a black arrow
plt.plot([v1[0],v3[0]],[v1[1],v3[1]],'b--') # plot v1 dashed blue line
plt.plot([v2[0],v3[0]],[v2[1],v3[1]],'r--') # plot v2 dashed red line
plt.plot([-6,6],[3,3],'k--') # plot x-axis
plt.plot([5,5],[-6,6],'k--') # plot x-axis
plt.legend()
plt.axis('square')
plt.axis((-6,6,-6,6))
plt.grid()
````

# Dot product
Dot product is single number.

It is defined as sum of products of corresponding elements of two vectors.

$ a \cdot b = \Sigma_{i=1}^n a_i b_i = a_1 b_1 + a_2 b_2 ... $

````python
v1 = np.array([1, 2  ,3,4,5,6])
v2 = np.array([0,-4,-3,6,5, 6])

dp1 = sum(np.multiply(v1,v2))

dp2 = np.dot(v1, v2)

# matrix multiplication
dp3 = np.matmul(v1,v2)

dp4 = 0
for i in range(0, len(v1)):
    dp4 = dp4 + v1[i]*v2[i]

print(dp1, dp2, dp3, dp4)
````

Geometric interpretation of dot product

$$ \alpha = \lVert a\rVert \lVert b \rVert cos(\theta)$$

$\theta$ is the angle between the vectors $a$ and $b$. If the angle is 90 degrees, then the dot product is 0.

$ cos \frac{\pi}{2} = 0 $

How to calculate angle between two vectors?
$$ cos(\theta) = \frac{a \cdot b}{\lVert a\rVert \lVert b \rVert} $$
$$ \theta = arcos (\frac{a \cdot b}{\lVert a\rVert \lVert b \rVert}) $$


Cauchy-Schwartz inequality
$$ |a^Tb | \le ||a|| \cdot ||b|| $$

````python
# Orthogonal vectors
v1 = np.array([0,2])
v2 = np.array([2,0])

plt.plot([0, v1[0]], [0,v1[1]], 'b', label='v_1')
plt.plot([0, v2[0]], [0,v2[1]], 'r', label='\lambda v_1')
plt.axis('square')
axlim = max([abs(max(v2)), abs(max(v1))]) * 1.5
plt.axis((-axlim, axlim, -axlim, axlim))
plt.grid()
````

## Distributive property
$$ a \cdot (b + c) = a \cdot b + a \cdot c $$


````python
n = 10
a = np.random.randint(n) # random integer between 0 and n-1
b = np.random.randint(n)
c = np.random.randint(n)

res1 = np.dot(a, b+c)
res2 = np.dot(a,b) + np.dot(a,c)

print([res1, res2])
````

## Associative property

$$ a \cdot (b \cdot c) = (a \cdot b) \cdot c $$


````python
## Associative property
n = 5
a = np.random.randint(low=1, high=100, size=n)
b = np.random.randint(low=1, high=100, size=n)
c = np.random.randint(low=1, high=100, size=n)

res1 = np.dot(a, np.dot(b,c))
res2 = np.dot(np.dot(a,b), c)

print(res1)
print(res2)
````

## Vector length
Norm, magnitude.
$$ ||v|| = \sqrt{v^Tv}$$

The length is simply the Pythagorean theorem.
$$ ||v||^2 = v^2_{1} + v^2_{2}$$

````python
v1 = np.array([1,2,3,4,5,6])

vl1 = np.sqrt(sum(np.multiply(v1,v1)))

vl2 = np.linalg.norm(v1)
print(vl1,vl2)
````

# Cross Product
Normal vector perpendicular to two vectors.

````python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

v1 = [ -3, 2, 5]
v2 = [ 4, -3, 0]
v3a = np.cross(v1,v2)

# manual method
v3b = [ [v1[1]*v2[2] - v1[2]*v2[1]],
        [v1[2]*v2[0] - v1[0]*v2[2]],
        [v1[0]*v2[1] - v1[1]*v2[0]]
]

fig = plt.figure()
ax = fig.add_subplot(projection='3d')


# draw the plane defined by span of v1 and v2
xx, yy = np.meshgrid(np.linspace(-10,10,10), np.linspace(-10,10,10))
z1 = (-v3a[0]*xx - v3a[1]*yy)/v3a[2]
ax.plot_surface(xx, yy, z1, color='blue', alpha=0.3)

#plot the vectors
ax.plot([0, v1[0]], [0, v1[1]], [0, v1[2]], 'g')
ax.plot([0, v2[0]], [0, v2[1]], [0, v2[2]], 'b')
ax.plot([0, v3a[0]], [0, v3a[1]], [0, v3a[2]], 'r')

ax.view_init(azim=150, elev=45)
plt.show()
````

# Hermitian transpose (conjugate transpose)
Conjugate of the complex number.
Flip the imaginary part

$a + bi = a - bi$ 

$a - bi = a + bi$

During the transpose you conjugate (flip) the imaginary part of complex number.

````python
z = np.array([3+4j])

#magnitude 
print(np.linalg.norm(z))

print(np.transpose(z.conjugate())*z)
````

# Unit vectors
Has lenght of 1.

$||\mu v || = 1$

You need reciprocical of vector v.
$ ||\mu v || = \frac{1}{||v||} ||v|| = 1$



````python
v1 = np.array([-3,6])

#mu
mu = 1/np.linalg.norm(v1)

v1n = v1*mu

# plot
plt.plot([0, v1n[0]], [0, v1n[1]], 'r', label='v1-norm', linewidth=5)
plt.plot([0, v1[0]], [0, v1[1]], 'b', label='v1')

plt.axis('square')
plt.axis((-6,6,-6,6))
plt.grid()
plt.legend()
````

# Subspace
Algebraic view - all linear combinations
$\lambda v$ any vector $v$ is in the same subspace.

v = [2 4] cannot become [2 4], they are not in the same subspace.

Subspace from two vectors
$\lambda v + \beta w$

Linear combination is scalar multiplication and addition

Formal definition: A vector subspace must
- be closed under addition and scalar multiplication
- contain the zero vector

# Subset
A set of points that satisfies some conditions
- Doesn't need to include origin
- Doesn't need to be closed
- Can have boundaries

Example (subset): All points on the XY plane such that x > 0, y > 0

Example (subset): All points on the XY plane such that $x^2 + y^2 = 1$

Example (subset and subspace): All points on the XY plane such that $y=4x$, all x

Example (subset): All points on the XY plane such that $y=4x+1$, all x

# Span

$span(\{v_1,v_2, ... , v_n\}) = \alpha_1v1_1 + ... + \alpha_nv_n, \alpha \in \mathbb{R}$

All possible linear combinations of set of vectors 

````python
# set S
S1 = np.array([1,1,0])
S2 = np.array([1,7,0])

# Vectors v and w
v = np.array([1,2,0])
w = np.array([3,2,1])

# draw vectors
fig = plt.figure()
ax = fig.gca(projection='3d')
ax.plot([0, S1[0]], [0, S1[1]], [.1, S1[2]+ .1], 'r', linewidth=3)
ax.plot([0, S2[0]], [0, S2[1]], [.1, S2[2]+ .1], 'r', linewidth=3)

ax.plot([0, v[0]], [0, v[1]], [.1, v[2]+ .1], 'g', linewidth=3)
ax.plot([0, w[0]], [0, w[1]], [0, w[2]], 'b')

# draw plane
xx, yy = np.meshgrid(range(-15,16), range(-15,16))
cp = np.cross(S1, S2)
Z1 = (-cp[0]*xx - cp[1]*yy)*1./cp[2]
ax.plot_surface(xx,yy, Z1)
````
