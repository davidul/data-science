# t-test
Main idea: testing if data drawn from distribution are significantly different from data drawn from another distribution.

$t_k = \frac{\bar{x} - \bar{y}}{\frac{s}{\sqrt{n}}} $

$\bar{x}, \bar{y}$ mean of x and y

$s$ standard deviation

$\frac{difference of means}{Standard deviations}$


Evaluate against distribution of t-values.

```python
import numpy as np
import matplotlib.pyplot as plt

import scipy.stats as stats
```

```python
n1 = 30 # samples in dataset 1
n2 = 40 # and 2
mu1 = 1 # population mean dataset 1
mu2 = 2 # population mean dataset 2

# generate data
data1 = mu1 + np.random.randn(n1)
data2 = mu2 + np.random.randn(n2)

# plot data
plt.plot(np.zeros(n1), data1, 'ro', markerfacecolor='w', markersize=14)
plt.plot(np.ones(n2), data2, 'bs', markerfacecolor='w', markersize=14)
plt.xlim([-1,2])
plt.xticks([0,1], labels=['Group 1', 'Group 2'])
```

```python
# t-test via stats package
t,p = stats.ttest_ind(data1, data2)
print(t)
print(p)
```

```python
fig = plt.figure(figsize=(10,4))
plt.rcParams.update({'font.size':12}) 

plt.plot(0+np.random.randn(n1)/15, data1, 'ro', markerfacecolor='w', markersize=14)
plt.plot(1+np.random.randn(n1)/15, data1, 'bs', markerfacecolor='w', markersize=14)

plt.xlim([-1,2])
plt.xticks([0,1], labels=['Group 1', 'Group 2'])

plt.title(f't = {t:.2f}, p={p:.3f}')
```
