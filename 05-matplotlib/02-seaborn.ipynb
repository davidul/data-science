# Distribution plots


```python
import seaborn as sns
import numpy as np
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
#sns.displot(tips['total_bill'], kde=False, bins=30)
sns.displot(tips['total_bill'], kde=True)
# kde kernel density estimation
```

```python
sns.displot(tips['total_bill'], kind='hist')
```

```python
sns.ecdfplot(tips['total_bill'])
# empirical cumulative distribution function
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips, kind='hex')
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips, kind='reg')
```

```python
sns.pairplot(data=tips, hue='sex')
```

```python
sns.rugplot(tips['total_bill'])
```

# Cat plot

```python
sns.catplot(data=tips, x='total_bill', y='sex')
```

```python
sns.catplot(data=tips, x='total_bill', y='sex', kind='box')
```

# Heat Map

```python
print("Smoker unique values",tips.smoker.unique())
print("Day unique values",tips.day.unique())
```

```python
tips.sex = tips.sex.map({'Female': 0, 'Male': 1})
tips.smoker = tips.smoker.map({'No': 0, 'Yes': 1})
tips.day = tips.day.map({'Thur': 0, 'Fri': 1, 'Sat': 2, 'Sun': 3})
tips.head()
```

```python
mask = np.triu(np.ones_like(tips.corr()))
sns.heatmap(tips.corr(), mask=mask,annot=True)
```
