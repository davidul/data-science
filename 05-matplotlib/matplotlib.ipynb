```python
import matplotlib.pyplot as plt
import numpy as np
```

Plots on [Figure](https://matplotlib.org/stable/api/figure_api.html#matplotlib.figure.Figure) and this object contains one or more [Axes](https://matplotlib.org/stable/api/_as_gen/matplotlib.axes.Axes.html#matplotlib.axes.Axes)

```python
fig, ax = plt.subplots()
ax.plot([1,2,3,4])
ax.set_title('Test')
ax.legend()
```

Plot x vs y

```python
x = np.linspace(0,2,100)

fig,ax = plt.subplots()

ax.plot(x,x, label='linear')
ax.plot(x,x**2, label='quadratic')
ax.plot(x,x**3, label='cubic')
ax.set_xlabel('x label')
ax.set_ylabel('y label')
ax.set_title('Title')
ax.legend()
```

```python
x = np.linspace(0,5,11)
y = x ** 2
```

```python
fig, ax = plt.subplots()
ax.plot(x,y, 'r--', x, y+2, 'bs')
```

```python
plt.subplot(1,2,1) # rows, cols, index -> 1row, 2cols, 1st index
plt.plot(x,y, 'r')

plt.subplot(1,2,2) # 1row, 2col, index
plt.plot(y,x,'b')
```

# OOP method

```python
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x,y)
axes.set_xlabel('X Label')
axes.set_ylabel('X Label')
axes.set_title('Title')
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,.4,.3])

axes1.plot(x,y)
axes2.plot(y,x)
axes2.set_title('INSERT')
```

```python
fig, axes = plt.subplots()
axes.plot(x,y)
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)
#axes.plot(x,y)
for current_axes in axes:
    current_axes.plot(x,y)
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)
axes[0].plot(x,y)
axes[0].set_title('One')
axes[1].plot(y,x)
axes[1].set_title('Two')
```

# Figure size and DPI

```python
fig = plt.figure(figsize=(8,2))
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
fix, axes = plt.subplots(figsize=(8,2))
axes.plot(x,y)
```

```python
fix, axes = plt.subplots(nrows=2, ncols=1, figsize=(8,2))
axes[0].plot(x,y)
axes[1].plot(y,x)
plt.tight_layout()
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x, x**2, label='Square')
ax.plot(x, x**3, label='Qube')

ax.legend(loc='upper right')
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y, color='orange', linewidth=3, alpha=0.5, linestyle='--', marker='*', markersize=15, markerfacecolor='blue',
        markeredgewidth=3, markeredgecolor='red')
#linewidth => lw
#linestyle => ls => '--' '-.' 'steps'
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
ax.set_xlim([0,1])
ax.set_ylim([0,2])
```
