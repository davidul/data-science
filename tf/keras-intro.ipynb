```python
import numpy as np
import tensorflow as tf
from tensorflow import keras
import matplotlib.pyplot as plt
```

# Activation Functions

## Relu

```python
c = tf.constant([-1,0,1, 2, 3, 4, 5, 6], dtype=tf.float64)
y = tf.keras.activations.relu(c, alpha=0.0, max_value=4, threshold=0.0).numpy()
```

```python
fig, ax = plt.subplots()
ax.plot(c,y)
```

## Sigmoid

```python
y = tf.keras.activations.sigmoid(c)
fig, ax = plt.subplots()
ax.plot(c,y)
```

## Softmax
Softmax converts a vector of values to a probability distribution.

```python
tf.reduce_sum()
inputs = tf.random.normal(shape=(32, 10))
y = tf.keras.activations.softmax(inputs)
fig, ax = plt.subplots()
ax.plot(inputs,y)
```
