# Min-max scaling

Scale range to 0 to 1
Formula: (x - min) / (max - min)
$$ \tilde{x} = \frac{x-min(x)}{max(x) - min(x)}$$

Scale to a range of a to b.
$$ x^* = a + \frac{(x - min(x))(b - a)}{max(x) - min(x)} = a + \tilde{x}(b-a)$$

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# create data

N = 42
data = np.log(np.random.rand(N)) * 235 + 934

data_min = data.min()
data_max = data.max()

data_scaled = (data - data_min) / (data_max - data_min)

fig, ax = plt.subplots(1, 2, figsize=(8, 4))
ax[0].plot(1+ np.random.rand(N)/20, data, 'ks')
ax[0].set_title('Original data')
ax[0].set_ylabel('Original data scala')
ax[0].set_xticks([])
ax[0].set_xlim(0, 2)

ax[1].plot(1+ np.random.rand(N)/20, data_scaled, 'ks')
ax[1].set_title('Scaled data')
ax[1].set_ylabel('Unity-normed data scala')
ax[1].set_xticks([])
ax[1].set_xlim(0, 2)

plt.show()
```

```python
## show that scaling doesn't affect the relative values
plt.plot(data, data_scaled, 'ks')
plt.xlabel('Original')
plt.ylabel('Scaled')
plt.show()
```

```python
## arbitrary scaling

newMin = 4
newMax = 8

dataSS = data_scaled * (newMax - newMin) + newMin
print(dataSS.min(), dataSS.max())
```
