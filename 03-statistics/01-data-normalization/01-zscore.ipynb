# Z-score

Z-score is the number of standard deviations from the mean a data point is.

$$z = \frac{x - \mu}{\sigma}$$

where $\mu$ is the population mean and $\sigma$ is the population standard deviation.

The Z-score is positive if the value lies above the mean, and negative if it lies below the mean.

The Z-score is zero if it lies on the mean.

The Z-score is a dimensionless quantity, and is useful for comparing data measured on different scales.

The Z-score is also known as the standard score.

The Z-score is a special case of the more general concept of a standard score.

Z-transform shift and stretches but does not change the shape of the distribution.

```python
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as stats
```

```python
## create a dataset

data = np.random.poisson(lam=3, size=1000)**2

datamean = np.mean(data)
datastd = np.std(data, ddof=1)

plt.plot(data, 's', markersize=3)
plt.xlabel('Data index')
plt.ylabel('Data value')
plt.title(f'Mean: {datamean:.2f}, Std: {datastd:.2f}')
plt.show()
```

```python
## z-score
dataz = (data - datamean) / datastd

dataz = stats.zscore(data)

dataZmean = np.mean(dataz)
dataZstd = np.std(dataz, ddof=1)

plt.plot(dataz, 's', markersize=3)
plt.xlabel('Data index')
plt.ylabel('Data value')
plt.title(f'Mean: {dataZmean:.2f}, Std: {dataZstd:.2f}')
plt.show()
```

```python
# show that the relative values are preserved

plt.plot(data, dataz, 's', markersize=3)
plt.xlabel('Original data')
plt.ylabel('Z-score')
plt.title(f'Correlation: {np.corrcoef(data, dataz)[0,1]:.2f}')
plt.show()
```
