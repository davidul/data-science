# Quintile-Quintile Plots

Check if data come from normal distribution.

```python
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as stats
```

```python
## generate data
n = 1000
data = np.random.randn(n)

# theoretical normal distribution
x = np.linspace(-4,4, 10001)
thenorm = stats.norm.pdf(x)
thenorm = thenorm/max(thenorm)

# plot histograms
yy, xx = np.histogram(data, 40)
yy = yy/max(yy)
xx = (xx[:-1] + xx[1:])/2

plt.plot(xx, yy, label='Empirical')
plt.plot(x, thenorm, label='Theoretical')
plt.legend()
plt.show()
```

```python
## QQ plot
zSortData = np.sort(stats.zscore(data))
sortNormal = stats.norm.ppf(np.linspace(0,1,n))

plt.plot(zSortData, 'o')
plt.plot(sortNormal, 'x')
```

```python

plt.plot(sortNormal, zSortData, 'o')

# set axes to be equal
xL, xR = plt.xlim()
yL, yR = plt.ylim()
lims = [ np.min([xL, xR, yL, yR]), np.max([xL, xR, yL, yR])]
plt.xlim(lims)
plt.ylim(lims)

plt.plot(lims, lims)
plt.xlabel('Theoretical Normal')
plt.ylabel('Observed data')
plt.title('QQ Plot')
```

```python
# Python solution

x = stats.probplot(data, plot=plt)
```

```python
# log-normal distribution
dataLogN = np.exp(np.random.randn(n)* .8)

x = stats.probplot(dataLogN, plot=plt)
```
