# Cumulative distribution functions

CDFs are useful for determining the probability that a random variable is less than or equal to a certain value.

PDF probability density function

CDF is the cumulative sum (integral) of the PDF (probability density function).

The y-axis of the CDF is the probability that the random variable is less than or equal to the corresponding x-axis value.

The y-axis value at each x-value is the sum of all probabilities of all x-values less than or equal to the x-value (to the left of the x-value).

CDF starts at 0 and increases monotonically to 1. The sum of CDF is more than 1.

$ C(x_a) = p(X \leq x_a) = \int_{-\infty}^{x_a} p(x) dx $

$ \Sigma_{i=1}^{a} p(x_a)$ 

```python
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as stats
```

```python
# example using log-normal distribution

# variable to evaluate the CDF

X = np.linspace(0, 5, 1001) # 1001 points between 0 and 5

p1 = stats.lognorm.pdf(X, 1) # log-normal with mean=1   and std=1
c1 = stats.lognorm.cdf(X, 1) # log-normal with mean=1   and std=1

p2 = stats.lognorm.pdf(X, 0.1) # log-normal with mean=0.1 and std=1
c2 = stats.lognorm.cdf(X, 0.1) # log-normal with mean=0.1 and std=1
```

```python
# plot the PDFs and CDFs

fig, ax = plt.subplots(2, 1, figsize=(4, 7))

ax[0].plot(X, p1, label='mean=1, std=1')
ax[0].plot(X, p2, label='mean=0.1, std=1')
ax[0].legend()
ax[0].set_title('PDF')

ax[1].plot(X, c1, label='mean=1, std=1')
ax[1].plot(X, c2, label='mean=0.1, std=1')
ax[1].legend()
ax[1].set_title('CDF')
plt.show()
```
