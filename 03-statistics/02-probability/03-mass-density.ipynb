# Probability mass vs density

Probability: Chance of an event occurring

Probability mass: A function that gives the probability that a discrete random variable is exactly equal to some value.

Probability Density: A function that describes probabilities of a continuous random variable.

## Probability mass function


```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# basic formula

# counts of different events
c = np.array([1, 2, 3, 4])

prob = c / c.sum()
print(prob)
print(100*prob)
```

```python
# example of drawing marbles from a bag

# colored marbles
blue = 40
yellow = 30
orange = 20
totalMarbs = blue + yellow + orange

jar = np.hstack((1*np.ones(blue), 2*np.ones(yellow), 3*np.ones(orange)))

# draw 500 marbles with replacement
numDraws = 500
drawColors = np.zeros(numDraws)

for i in range(numDraws):
    drawColors[i] = jar[np.random.randint(0, totalMarbs)]


propBlue = np.sum(drawColors == 1) / numDraws
propYellow = np.sum(drawColors == 2) / numDraws
propOrange = np.sum(drawColors == 3) / numDraws

# plot the results
plt.bar([1, 2, 3], [propBlue, propYellow, propOrange], label='Proportion')
plt.plot([0.5, 1.5], [blue/totalMarbs, blue/totalMarbs], 'b', label='True probability')
plt.plot([1.5, 2.5], [yellow/totalMarbs, yellow/totalMarbs], 'y')
plt.plot([2.5, 3.5], [orange/totalMarbs, orange/totalMarbs], 'orange')

plt.xticks([1, 2, 3], ['Blue', 'Yellow', 'Orange'])
plt.xlabel('Marble Color')
plt.ylabel('Proportion')
plt.legend()
plt.show()
```

## Probability density function
Continuous random variable: A random variable that can take any value in a given range.

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# compute empirical probability function

# continuos signal (technically discrete
N = 10004
datats1 = np.cumsum(np.sign(np.random.randn(N)))
datats2 = np.cumsum(np.sign(np.random.randn(N)))

# plot the results
plt.plot(np.arange(N), datats1, linewidth=2)
plt.plot(np.arange(N), datats2, linewidth=2)
plt.show()

# discretize using histogram
nbins = 50
y,x = np.histogram(datats1, nbins)
x1 =(x[1:]+x[:-1])/2
y1 = y/np.sum(y)

y,x = np.histogram(datats2, nbins)
x2 =(x[1:]+x[:-1])/2
y2 = y/np.sum(y)

# density plot
plt.plot(x1, y1, x2, y2, linewidth=3)
plt.legend(['ts1', 'ts2'])
plt.xlabel('Data Value')
plt.ylabel('Probability')
plt.show()
```
