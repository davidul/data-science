# Normal distribution

$$ f(x) = \frac{1}{\sigma \sqrt{2\pi}} e^{-\frac{1}{2} (\frac{x-\mu}{\sigma})^2} $$

The core of the above is $ e^{-x^2}$ .
Where $x$ is the distance from the mean, and $\sigma$ is the standard deviation.

Normal distribution is a continuous probability distribution. It is symmetric about the mean, showing that data near the mean are more frequent in occurrence than data far from the mean.

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-2,2, 1000)
f = np.exp(-x**2)

plt.plot(x, f)
```
