# Violin plots

Violin plots are similar to box plots, except that they also show the probability density of the data at different values, usually smoothed by a kernel density estimator. Denser regions of the data are fatter, and sparser thiner in a violin plot


```python
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as stats
```

```python
## Create the data

n = 1000
threshold = 5

data = np.exp(np.random.randn(n))
data[data > threshold] = threshold + np.random.randn((data > threshold).sum()) * 0.1

plt.hist(data, bins=30)
plt.title('Histogram of data')
plt.show()

# show the violin plot
plt.violinplot(data, showmeans=False, showmedians=True)
plt.title('Violin plot of data')
plt.show()
```
