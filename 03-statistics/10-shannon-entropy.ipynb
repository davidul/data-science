# Shannon entropy

Formula, x = data values, p(x) = probability of x

For log2, the unit is bits. For ln (natural log), the unit is nats.

$$ H = - \Sigma_{i=1}^{n} p(x_i) log_2(p(x_i))$$

High entropy = high uncertainty, means dataset has a lot of variability.

Low entropy = low uncertainty, means dataset has less variability. Most of the data repeats.

How does entropy differs from variance?

Entropy is nonlinear and makes no assumption about the distribution of the data. Variance is linear and assumes the data is normally distributed.



```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
## discrete entropy

N = 1000
numbers = np.ceil(8*np.random.rand(N)**2)

u = np.unique(numbers)
probs = np.zeros(len(u))

for i in range(len(u)):
    probs[i] = np.sum(numbers == u[i]) / N

entropy = -np.sum(probs * np.log2(probs))

plt.bar(u, probs)
plt.xlabel('number')
plt.ylabel('probability')
plt.title('entropy = %.2f' % entropy)
plt.show()
```

```python
## for random variables
# create brownian motion

N = 1123
brownian = np.cumsum(np.sign(np.random.randn(N)))

fig,ax = plt.subplots(2,1, figsize=(4, 6))
ax[0].plot(brownian)
ax[0].set_xlabel('Data Index')
ax[0].set_ylabel('Value')
ax[0].set_title('Brownian Motion')

ax[1].hist(brownian/sum(brownian), bins=30)
ax[1].set_xlabel('Value')
ax[1].set_ylabel('Probability')
ax[1].set_title('Histogram of Brownian Motion')
plt.show()
```

```python
## compute entropy
nbins = 50
counts, bins = np.histogram(brownian, bins=nbins)
probs = counts / sum(counts)
entropy = -np.sum(probs * np.log2(probs))
print('Entropy of Brownian Motion = %.2f' % entropy)
```
