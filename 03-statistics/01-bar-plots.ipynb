# Bar plots

bar(x, height, width=0.8, bottom=None, *, align='center', data=None, **kwargs)
- x - defines the position on x axis
- height - defines the height of the bar
- width - defines the width of the bar
- bottom - defines the bottom of the bar

$ (\frac{2i}{n-1} - 1)^2 + (i+1)^2$

Plot the $(i+1)^2$

````python
x = np.arange(0, 10, 1)
y = (x + 1) ** 2
plt.plot(x,y)
plt.show()
````

Plot $ (\frac{2i}{n-1} - 1)^2 $

````python
x = np.arange(0, 10, 1)
y = ((2*x)/(len(x)-1) - 1)**2
plt.plot(x,y)
plt.show()
````

````python
m = 30
n = 6

# matrix of zeros m x n
data = np.zeros((m,n))

# (i+2)**2 scale term, every element is greater then the previous
# (2*i/(n-1)-1)**2 variance term
for i in range(n):
    data[:,i] = 30*np.random.randn(m) * (2*i/(n-1)-1)**2 + (i+1)**2
````

````python
data
````

````python
fig, ax = plt.subplots(1,3,figsize=(8,3))

# bars
ax[0].bar(range(n), np.mean(data, axis=0))
ax[0].set_title('Bar plot')

#error bars
ax[1].errorbar(range(n), np.mean(data, axis=0), np.std(data, axis=0, ddof=1), marker='s', linestyle='')
ax[1].set_title('Error bar plot')

# both
ax[2].bar(range(n), np.mean(data, axis=0))
ax[2].errorbar(range(n), np.mean(data, axis=0), np.std(data, axis=0, ddof=1), marker='s', linestyle='', color='r')
ax[2].set_title('Error+bar plot')
````

````python
# manually specify x-axis coordinates
xcrossings = [1,2,4,5,6,9]

plt.bar(xcrossings, np.mean(data, axis=0))
plt.errorbar(xcrossings, np.mean(data,axis=0), np.std(data, axis=0, ddof=2), marker='.',linestyle='', color='b')
plt.title('Bar error plot')
````

````python
# bars from matrices

m = [ [2,5,4,3], [1,1,8,6]]
fig,ax = plt.subplots(nrows=2, ncols=2, figsize=(8,8))
ax[0,0].imshow(m)

df = pd.DataFrame(m, columns=['prop 0', 'prop 1', 'prop 2', 'prop 3'])
df.plot(ax=ax[1,0], kind='bar')
ax[1,0].set_title('Grouping by rows')

ax[0,1].imshow(np.array(m).T)
df.T.plot(ax=ax[1,1],kind='bar')
ax[1,1].set_title('Grouping by columns')
````
