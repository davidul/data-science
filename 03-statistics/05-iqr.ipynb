# Inter Quartile Range
Median is quartile 2.

Quartile 1 and quartile 3 are left/right from median.

50% of data is between quartile 1 and quartile 3.

```python
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as stats
```

```python
## Create the data
n = 1000
data = np.random.randn(n)**2

plt.plot(data, 'p')
plt.show()
plt.hist(data)
plt.show()
```

```python
data[:5]
```

```python

# rank transform the data and scale to 1
ranked_data = stats.rankdata(data)
dataR = stats.rankdata(data)/n 
plt.plot(dataR, 's')
print(ranked_data[:5])
print(dataR[:5])
```

```python
q1 = np.argmin((dataR-.25)**2)
q3 = np.argmin((dataR-.75)**2)

iq_vals = data[[q1,q3]]

iqrange1 = iq_vals[1] - iq_vals[0]

iqrange2 = stats.iqr(data) # interquartile range

print(iqrange1, iqrange2)
```
