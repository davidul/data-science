# Inter Quartile Range

The interquartile range (IQR) is a measure of statistical dispersion. It is defined as the difference between the 75th (Q3) and 25th (Q1) percentiles of the data. The IQR is a robust measure of dispersion, meaning that it is not as sensitive to outliers as some other measures, such as the range.

The third quartile is the median of the upper half of the data set, and the first quartile is the median of the lower half of the data set.

Median is quartile 2.

50% of data is between quartile 1 and quartile 3.

The IQR can be calculated using the following steps:

1. Order the data from least to greatest.
2. Find the median, which is the middle value in the ordered data set.
3. Find the first quartile (Q1), which is the median of the lower half of the data set.
4. Find the third quartile (Q3), which is the median of the upper half of the data set.
5. Calculate the IQR by subtracting Q1 from Q3.

The IQR can be used to identify outliers in a data set. A data point is considered an outlier if it is more than 1.5 IQRs below Q1 or above Q3.

The IQR can also be used to compare the spread of data sets. For example, if two data sets have the same median, but one data set has a larger IQR, then that data set has a wider spread of data points.

```python
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as stats
```

```python
## Create the data
n = 1000
data = np.random.randn(n)**2

plt.plot(data, 'p')
plt.show()
plt.hist(data)
plt.show()
```

```python
data[:5]
```

```python
q1, q2 = np.percentile(data, [25,75])
print(q1, q2)
```

```python

# rank transform the data and scale to 1
ranked_data = stats.rankdata(data)
dataR = stats.rankdata(data)/n 
plt.plot(dataR, 's')
print(ranked_data[:5])
print(dataR[:5])
```

```python
(dataR-.25)
```

```python
q1 = np.argmin((dataR-.25)**2)
q3 = np.argmin((dataR-.75)**2)

iq_vals = data[[q1,q3]]

iqrange1 = iq_vals[1] - iq_vals[0]

iqrange2 = stats.iqr(data) # interquartile range

print(iqrange1, iqrange2)
```
